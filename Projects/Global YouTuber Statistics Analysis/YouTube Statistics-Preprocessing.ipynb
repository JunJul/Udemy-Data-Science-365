## Missing values

```python
interested_data.isnull().sum()
```

```python
noNull_data = interested_data.dropna(axis=0)
noNull_data.isnull().sum()
```

```python
noNull_data = noNull_data.reset_index(drop=True)
```

## Outliers

```python
import matplotlib.pyplot as plt
for i in noNull_data.columns:
    if i != 'category':
        sns.distplot(noNull_data[i])
        plt.show()
```

```python
def dealing_outliers(df, colName, min_q, max_q):
    for i in colName:
        min_thresold, max_thresold = df[i].quantile([min_q, max_q])
        df = df[(df[i] > min_thresold) & (df[i] < max_thresold)]
    return df
```

```python
quantiled_columns = list(('video views', 'uploads', 'video_views_for_the_last_30_days',
    'highest_yearly_earnings', 'subscribers_for_last_30_days', 'created_year'))
noNull_outliers_data = dealing_outliers(noNull_data.copy(), quantiled_columns, 0.05, 0.95)
```

```python
noNull_outliers_data.describe()
```

## Checking linearity

```python
def checkingLinearity(df, columns, y_columns):
    plt.figure(figsize=(5,3))
    y = df[y_columns]
    for i in range(len(columns)):
        x = df[columns[i]]
        plt.scatter(x, y)
        plt.xlabel(x.name)
        plt.ylabel(y.name)
        plt.show()
```

```python
checking_linearity_columns = ['subscribers', 'video views', 'uploads', 'video_views_for_the_last_30_days',
                              'subscribers_for_last_30_days', 'created_year']
checkingLinearity(noNull_outliers_data, checking_linearity_columns, 'highest_yearly_earnings')
```

## Get dummies

```python
dummies_data = pd.get_dummies(noNull_outliers_data, drop_first=True, dtype=int)
dummies_data
```

## Checking multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable = dummies_data.drop('highest_yearly_earnings', axis=1)
```

```python
vif = [variance_inflation_factor(variable.values, i) for i in range(len(variable.columns))]
vif_data_table = pd.DataFrame({'Feature': variable.columns, 'VIF': vif})
vif_data_table
```

```python
variable = variable.drop(['created_year','uploads'], axis=1)
vif = [variance_inflation_factor(variable.values, i) for i in range(len(variable.columns))]
vif_data_table = pd.DataFrame({'Feature': variable.columns, 'VIF': vif})
vif_data_table
```

```python
cols = variable.columns
cols = cols.insert(0, 'highest_yearly_earnings')
data_preprocessed = dummies_data[cols].reset_index(drop=True)
data_preprocessed
```

## get input and target

```python
unscaled_data_inputs = data_preprocessed.iloc[:, 1:]
data_targets = data_preprocessed.iloc[:, :1]
```

## feature selection

```python
from sklearn.feature_selection import f_regression
```

```python
p_values = f_regression(unscaled_data_inputs, data_targets)
p_values = p_values[1].round(3)

p_values_table = pd.DataFrame({'Features': unscaled_data_inputs.columns, 'P_Values': p_values})
p_values_table
```

```python
cols = unscaled_data_inputs.columns[:4]
cols
```

```python
unscaled_data_feature_selected_inputs = unscaled_data_inputs[cols].copy()
```

## Adjusted R^2

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(unscaled_data_feature_selected_inputs, data_targets)
r2 = linear_reg.score(unscaled_data_feature_selected_inputs, data_targets)
```

```python
n = unscaled_data_feature_selected_inputs.shape[0]
p = unscaled_data_feature_selected_inputs.shape[1]
```

```python
adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
print(r2, adjusted_r2)
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_input_data = scaler.fit_transform(unscaled_data_feature_selected_inputs)
scaled_input_data
```

## Splits

```python
total_sample_counts = unscaled_data_feature_selected_inputs.shape[0]

train_sample_counts = int(0.8*total_sample_counts)
validation_sample_counts = int(0.1*total_sample_counts)
test_sample_counts = total_sample_counts - train_sample_counts - validation_sample_counts
```

```python
print(total_sample_counts, 
      train_sample_counts, validation_sample_counts, test_sample_counts)
```

```python
train_input = scaled_input_data[: train_sample_counts]
train_target = data_targets[: train_sample_counts]
```

```python
validation_input = scaled_input_data[train_sample_counts: train_sample_counts+validation_sample_counts]
validation_target = data_targets[train_sample_counts: train_sample_counts+validation_sample_counts]
```

```python
test_input = scaled_input_data[train_sample_counts+validation_sample_counts: ]
test_target = data_targets[train_sample_counts+validation_sample_counts: ]
```

## Save to npz

```python
np.savez('train_data', inputs=train_input, targets=train_target)
np.savez('validation_data', inputs=validation_input, targets=validation_target)
np.savez('test_data', inputs=test_input, targets=test_target)
```
